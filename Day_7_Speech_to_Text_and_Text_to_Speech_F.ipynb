```python
import speech_recognition as spr
import soundfile
import librosa
from gtts import gTTS
import IPython.display as ipd
```

```python
recognizer = spr.Recognizer()
```

```python
ipd.Audio('NLPSr.mp3')
```

```python
x,sr = librosa.load('NLPSr.mp3')
```

```python
soundfile.write('NLPSr.wav',
                data = x,
                samplerate =sr,
                subtype='PCM_24')
```

```python
ipd.Audio('NLPSr.wav')
```

```python
with spr.AudioFile('NLPSr.wav') as source:
    audio = recognizer.record(source)
```

```python
recognizer.recognize_google(audio)
```

```python
from gtts import gTTS
```

```python
t_text = "To bring about any effective change, we need to rethink development keeping in mind that people are important, but equally important is the planet on which the prosperity of the people depends. Our civilisation demands that wars should always be moved from battlefields to peace tables. More importantly, the higher human values that spring from the awareness of the spiritual basis of our being, must be reawakened if any crisis is to be resolved."
```

```python
newobj = gTTS(t_text, lang='en', tld='com', slow=False)
```

```python
newobj.save('Synthspeech.mp3')
```

```python

ipd.Audio('Synthspeech.mp3')
```

```python
gTTS(t_text, lang='en', tld='co.in', slow=False).save('SynthSpeech_in.mp3')
```

```python
ipd.Audio('SynthSpeech_in.mp3')
```
