```python
import warnings
warnings.filterwarnings('ignore')

from fuzzywuzzy import process
```

```python
ipl_teams = ['Chennai Super Kings',
             'Royal Challengers Bangalore',
             'Delhi Capitals',
             'Kolkata Knight Riders',
             'Punjab Kings',
             'Rajasthan Royals',
             'Sunrisers Hyderabad']
```

```python
#Play -> #Stay
# First remove P and then insert S in place of P (2 operations)
# Remove l and insert t (2 operations)
```

```python
statement = 'Kolkata Knights beat dilli team by 30 runs'
process.extract(statement, ipl_teams, limit=2)
```
